## Asignment 1 : We have the min and max temperatures in a city In India for each months of the year. We would like to find a function to describe this and show it graphically, the dataset given below.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import optimize
```

```python
Max = np.array([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])
Min = np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])
months = np.arange(12)
```

```python
#defining periodic function
def yearly_temps(times, avg, ampl, time_offset):
    return (avg + ampl * np.cos((times + time_offset) * 1.8 * np.pi / times.max()))
#identifying the curve for plot using scipy
res_max, cov_max = optimize.curve_fit(yearly_temps, months,Max, [40, 20,0])
res_min, cov_min = optimize.curve_fit(yearly_temps, months,Min, [-40, 20,0])
```

```python
#2. plot the fit
days = np.linspace(0, 12, num=365)
plt.figure()
plt.plot(months,Max, 'ro')
plt.plot(days, yearly_temps(days, *res_max), 'r-')
plt.plot(months,Min, 'bo')
plt.plot(days, yearly_temps(days, *res_min), 'b-')
plt.xlabel('Month')
plt.ylabel('Temperature (°C)')
plt.suptitle('function to describe this and show it graphically', fontsize = 15)
plt.show()
```

## Assignment 2: This assignment is for visualization using matplotlib

```python
url='https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv'
data = pd.read_csv(url)
```

```python
data.head()
```

### 1. Create a pie chart presenting the male/female proportion

```python
sex_ratio= (data['sex'].value_counts(1)*100).round(2)
pd.DataFrame(sex_ratio)
```

```python
data['sex'].value_counts().plot(kind = 'pie', explode = [0, 0.1], autopct = '%.2f%%',startangle = 90,
                                       labels = ['Male','Female'],
                                         shadow = True, pctdistance = 0.5)
plt.axis('off')
plt.suptitle('pie chart presenting the male/female proportion', fontsize = 15)
plt.show()
```

### 2. Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

```python
category1 = data[data.sex=='male'].plot.scatter('age', 'fare', color='blue',label='male')
data[data.sex=='female'].plot.scatter('age', 'fare',color='red',label='female',ax=category1)
plt.suptitle('scatterplot with the Fare paid and the Age, differ the plot color by gender', fontsize = 15)
plt.show()
```
